**Gráfico 1: Víctimas por día de la Semana**

El gráfico que se muestra representa la cantidad de víctimas agrupadas por día de la semana. Cada barra en el gráfico corresponde a un día de la semana, y la altura de cada barra indica la cantidad total de víctimas ocurridas en ese día durante un período de tiempo específico.

Las etiquetas en el eje x indican los días de la semana, y el eje y muestra la cantidad de víctimas.

Este gráfico de barras es útil para visualizar patrones en la distribución de víctimas a lo largo de la semana. Puede ayudar a identificar si hay días específicos en los que ocurren más incidentes con víctimas o si hay algún patrón semanal en la cantidad de víctima

```python
# Agrupar los datos y calcular la suma de víctimas por día de la semana
victimas_por_dia_de_semana = df.groupby('day_of_week')['casualties'].sum().reset_index()

# Diccionario para traducir los días de la semana
dias_de_la_semana = {
    'Monday': 'Lunes',
    'Tuesday': 'Martes',
    'Wednesday': 'Miércoles',
    'Thursday': 'Jueves',
    'Friday': 'Viernes',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}

# Traducir los días de la semana en el DataFrame
victimas_por_dia_de_semana['day_of_week'] = victimas_por_dia_de_semana['day_of_week'].map(dias_de_la_semana)

# Crear un gráfico de barras con estilo mejorado
plt.figure(figsize=(10, 6))
plt.bar(victimas_por_dia_de_semana['day_of_week'], victimas_por_dia_de_semana['casualties'], color='#1f77b4', alpha=0.7, edgecolor='black')

# Establecer el título del gráfico
plt.title('Víctimas por Día de la Semana', fontsize=16)

# Etiquetas de los ejes y tamaño de fuente
plt.xlabel('Día de la Semana', fontsize=12)
plt.ylabel('Víctimas', fontsize=12)

# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45, fontsize=10)

# Añadir etiquetas en las barras
for x, y in zip(victimas_por_dia_de_semana['day_of_week'], victimas_por_dia_de_semana['casualties']):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10)

# Mostrar el gráfico
plt.tight_layout()
plt.show()

```

El gráfico representa la cantidad de víctimas a lo largo de los años y ofrece dos visualizaciones diferentes de estos datos:

**Gráfico 2: Gráfico de Barras**

En la parte superior del gráfico se puede visualizar una serie de barras verticales, una por cada año en el conjunto de datos.
Cada barra representa un año específico y su altura indica la cantidad total de víctimas que ocurrieron en ese año.
La línea horizontal punteada de color gris representa el promedio de víctimas en todos los años. Esto proporciona una referencia para comparar cada año con la media.
El eje x muestra los años, y el eje y representa la cantidad de víctimas.

**Gráfico 3: Histograma**

En la parte inferior del gráfico, se muestra un histograma que representa la distribución de la cantidad de víctimas a lo largo de los años.
El eje x muestra la cantidad de víctimas en intervalos, mientras que el eje y muestra la frecuencia con la que ocurren esas cantidades de víctimas.
El histograma proporciona una visión más detallada de cómo se distribuyen las víctimas a lo largo de los años y muestra si existe alguna tendencia o patrón en la frecuencia de las víctimas.
Estos gráficos ayudan a entender la variación en el número de víctimas a lo largo de los años y si hay años que se destacan en términos de incidentes. El promedio proporciona una referencia para evaluar si un año en particular tuvo una cantidad de víctimas por encima o por debajo de la media.







```python
# Agrupar los datos y calcular la suma de víctimas por año
casualties_by_year = df.groupby('year')['casualties'].sum().reset_index()
mean_casualties = casualties_by_year['casualties'].mean()

# Colores y estilo estético
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
plt.figure(figsize=(10, 6))

# Crear un gráfico de barras con una línea horizontal para el promedio
bars = plt.bar(casualties_by_year['year'], casualties_by_year['casualties'], color=colors)
plt.axhline(y=mean_casualties, color='grey', linestyle='--', label=f"Promedio: {mean_casualties:.2f}")

# Estilo estético de las barras
for bar in bars:
    bar.set_edgecolor('black')

plt.title('Víctimas a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Víctimas')
plt.legend()

# Mostrar el gráfico de barras
plt.show()

# Crear un histograma de los datos
plt.figure(figsize=(10, 6))

# Estilo estético del histograma
plt.hist(casualties_by_year['casualties'], bins=10, edgecolor='black', color='#1f77b4', alpha=0.7)
plt.title('Víctimas a lo largo de los Años (Histograma)')
plt.xlabel('Víctimas')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.show()

```

**Gráfico 4:**

En el eje horizontal (eje x) se muestra la cantidad total de víctimas de incidentes violentos en cada estado.


**Gráfico 5:**

El gráfico de pastel representa la distribución de víctimas de incidentes violentos según diferentes tipos de tiroteo.


El círculo se divide en segmentos, cada uno etiquetado con un tipo de tiroteo.

Cada segmento del pastel muestra la proporción de víctimas que se atribuyen a un tipo de tiroteo específico en relación con el total de víctimas.

Las etiquetas dentro de cada segmento muestran el nombre del tipo de tiroteo y el porcentaje de víctimas que representa en el total.

Dicho gráfico puuede ayudar a identificar cuáles son los tipos de tiroteo más comunes en términos de víctimas y proporcionar una comprensión general de la proporción de víctimas en cada categoría de tiroteo.







```python

# Agrupar los datos y calcular la suma de víctimas por tipo de tiroteo
victimas_por_tipo_de_tiroteo = df.groupby('shooting_type')['casualties'].sum().reset_index()

# Filtrar los tipos de tiroteo con víctimas
victimas_por_tipo_de_tiroteo = victimas_por_tipo_de_tiroteo[victimas_por_tipo_de_tiroteo['casualties'] != 0]

# Traducir etiquetas al español
traduccion_tipos_tiroteo = {
    'indiscriminate': 'Indiscriminado',
    'targeted': 'Dirigido',
    'accidental or targeted': 'Accidental o Dirigido',
    'accidental': 'Accidental',
    'targeted and indiscriminate': 'Dirigido e Indiscriminado',
    'hostage suicide': 'Secuestro y Suicidio',
    'public suicide (attempted)': 'Suicidio Público (Intento)',
    'public suicide': 'Suicidio Público',
    'unclear': 'Dudoso',
    'suicide': 'Suicidio'
}

victimas_por_tipo_de_tiroteo['shooting_type'] = victimas_por_tipo_de_tiroteo['shooting_type'].map(traduccion_tipos_tiroteo)

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Colores personalizados para el gráfico de pastel
colores = sns.color_palette('Pastel1')

# Crear un gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(victimas_por_tipo_de_tiroteo['casualties'], labels=victimas_por_tipo_de_tiroteo['shooting_type'], autopct='%1.1f%%', startangle=140, colors=colores, wedgeprops={'edgecolor': 'white'})

# Configurar el título del gráfico
plt.title('Víctimas por Tipo de Tiroteo', fontsize=16)

# Mostrar el gráfico
plt.axis('equal')  # Para asegurar que el gráfico de pastel sea un círculo
plt.show()

```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress


tiroteos_privadas = df[df['school_type'] == 'private']
tiroteos_publicas = df[df['school_type'] == 'public']


tiroteos_privadas_por_año = tiroteos_privadas.groupby('year').size()
tiroteos_publicas_por_año = tiroteos_publicas.groupby('year').size()

slope_privadas, intercept_privadas, _, _, _ = linregress(tiroteos_privadas_por_año.index, tiroteos_privadas_por_año.values)


slope_publicas, intercept_publicas, _, _, _ = linregress(tiroteos_publicas_por_año.index, tiroteos_publicas_por_año.values)


plt.figure(figsize=(8, 6))


plt.scatter(tiroteos_privadas_por_año.index, tiroteos_privadas_por_año.values, color='#1f77b4', label='Escuelas Privadas', s=100, alpha=0.7)


plt.scatter(tiroteos_publicas_por_año.index, tiroteos_publicas_por_año.values, color='#ff7f0e', label='Escuelas Públicas', s=100, alpha=0.7)


plt.plot(tiroteos_privadas_por_año.index, slope_privadas * tiroteos_privadas_por_año.index + intercept_privadas, color='#1f77b4', linestyle='--', linewidth=2, label='Regresión Privadas')


plt.plot(tiroteos_publicas_por_año.index, slope_publicas * tiroteos_publicas_por_año.index + intercept_publicas, color='#ff7f0e', linestyle='--', linewidth=2, label='Regresión Públicas')

plt.title('Relación entre Tiroteos en Escuelas Privadas y Públicas a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Cantidad de Tiroteos')
plt.legend()


plt.show()

print(f"Regresión para escuelas privadas: y = {slope_privadas:.2f}x + {intercept_privadas:.2f}")
print(f"Regresión para escuelas públicas: y = {slope_publicas:.2f}x + {intercept_publicas:.2f}")
```

```python
from scipy.stats import pearsonr

corr_privadas, _ = pearsonr(tiroteos_privadas_por_año.index, tiroteos_privadas_por_año.values)
print(f"Coeficiente de Pearson para escuelas privadas: {corr_privadas:.2f}")

corr_publicas, _ = pearsonr(tiroteos_publicas_por_año.index, tiroteos_publicas_por_año.values)
print(f"Coeficiente de Pearson para escuelas públicas: {corr_publicas:.2f}")
```

**Escuelas privadas:** El coeficiente de correlación de Pearson es 0.46 para las escuelas privadas. Esto indica una correlación moderada positiva entre los años y la cantidad de tiroteos en escuelas privadas. En otras palabras, hay una tendencia ascendente en la cantidad de tiroteos a lo largo de los años, pero la relación no es extremadamente fuerte.

**Escuelas públicas:** El coeficiente de correlación de Pearson es 0.6 para las escuelas públicas. Esto indica una correlación moderadamente fuerte y positiva entre los años y la cantidad de tiroteos en escuelas públicas. La relación es más fuerte que en el caso de las escuelas privadas, lo que sugiere una tendencia más marcada en el aumento de la cantidad de tiroteos a lo largo de los años en escuelas públicas.

```python
public_school_data = df[df['school_type'] == 'public']
private_school_data = df[df['school_type'] == 'private']


public_casualties_by_shooting_type = public_school_data.groupby('shooting_type')['casualties'].sum().reset_index()
private_casualties_by_shooting_type = private_school_data.groupby('shooting_type')['casualties'].sum().reset_index()


fig, axes = plt.subplots(1, 2, figsize=(18, 9))


num_slices_public = len(public_casualties_by_shooting_type)
num_slices_private = len(private_casualties_by_shooting_type)


#create a 3-d effect on pie-slices
explode_public = tuple([0.23] * num_slices_public)
explode_private = tuple([0.13] * num_slices_private)


public_colors = sns.color_palette("viridis", num_slices_public)
private_colors = sns.color_palette("inferno", num_slices_private)


wedges_public, _, autotexts_public = axes[0].pie(
    public_casualties_by_shooting_type['casualties'],
    labels=None,
    startangle=90,
    shadow=True,
    autopct='',
    explode=explode_public,
    colors=public_colors,
)


axes[0].axis('equal')
axes[0].set_title('Total Casualties by Shooting Type in PUBLIC Schools')


wedges_private, _, autotexts_private = axes[1].pie(
    private_casualties_by_shooting_type['casualties'],
    labels=None,
    startangle=90,
    shadow=True,
    autopct='',
    explode=explode_private,
    colors=private_colors,
)


axes[1].axis('equal')
axes[1].set_title('Total Casualties by Shooting Type in PRIVATE Schools')


legend_labels_public = [
    f"{label}: {pct:.1f}%"
    for label, pct in zip(public_casualties_by_shooting_type['shooting_type'], public_casualties_by_shooting_type['casualties'] / public_casualties_by_shooting_type['casualties'].sum() * 100)
]

legend_labels_private = [
    f"{label}: {pct:.1f}%"
    for label, pct in zip(private_casualties_by_shooting_type['shooting_type'], private_casualties_by_shooting_type['casualties'] / private_casualties_by_shooting_type['casualties'].sum() * 100)
]


axes[0].legend(wedges_public, legend_labels_public, title="Shooting Type", loc="upper left", bbox_to_anchor=(-0.05,-0.05))
axes[0].axis('off')


axes[1].legend(wedges_private, legend_labels_private, title="Shooting Type", loc="upper left", bbox_to_anchor=(-0.05,-0.05))
axes[1].axis('off')


fig.suptitle("Total Casualties by Shooting Type in Public and Private Schools", fontsize=16)


plt.show()
```

```python
import statsmodels.api as sm
import pandas as pd

def forward_selection(data, target, significance_level=0.2):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

**Preparación de datos:** Se seleccionaron las características numéricas del dataset, excluyendo la columna 'killed' como variable objetivo (y).

**Forward Selection:** Se implementó el método de Forward Selection para seleccionar las mejores características que tienen un impacto significativo en la variable objetivo 'killed'. El proceso consistió en evaluar iterativamente las características restantes y agregar aquellas cuyo valor p (p-value) fue menor que el nivel de significancia establecido (0.2 en este caso). El proceso continuó hasta que ninguna característica adicional cumplió con el criterio de significancia.

**Resultados:** Las características seleccionadas como más relevantes para predecir el número de personas asesinadas en tiroteos en escuelas fueron 'casualties' (total de víctimas), 'injured' (número de personas heridas) y 'two_or_more' (indicador de si hubo dos o más atacantes).

**Backward Elimination**



```python
def backward_elimination(data, target,significance_level = 0.2):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break
    return features
```

```python
backward_elimination(X,y)
```

**Preparación de datos:** Se utilizaron todas las características numéricas del dataset, excluyendo la columna 'killed' como variable objetivo (y).

**Backward Elimination:** Se implementó el método de Backward Elimination para eliminar iterativamente las características menos significativas. En cada iteración, se ajustó un modelo de regresión lineal utilizando todas las características disponibles y se evaluaron los valores p asociados a cada característica. Si el valor p de la característica con el valor p máximo era mayor o igual al nivel de significancia establecido (0.2 en este caso), esa característica se eliminaba del conjunto de características. Este proceso continuó hasta que todas las características restantes tuvieron un valor p por debajo del nivel de significancia.

**Resultados:** Las características seleccionadas como más relevantes para predecir el número de personas asesinadas en tiroteos en escuelas utilizando Backward Elimination fueron 'injured' (número de personas heridas), 'casualties' (total de víctimas) y 'two_or_more' (indicador de si hubo dos o más atacantes).

En resumen, estas características fueron identificadas como las más importantes para entender y predecir el número de personas asesinadas en tiroteos escolares utilizando el método de Backward Elimination

```python
def stepwise_selection(data, target,SL_in=0.1,SL_out = 0.1):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break
        else:
            break
    return best_features
```

```python
stepwise_selection(X,y)
```

**Preparación de Datos:** Se utilizaron todas las características numéricas del dataset, excluyendo la columna 'killed' como variable objetivo (y).

**Stepwise Selection:** Se implementó el método de Stepwise Selection, que incluye dos etapas:

**Forward Selection (Adición de Características):** Se agregaron iterativamente características al modelo si el valor p asociado a esa característica era menor que el nivel de significancia de entrada (SL_in, establecido en 0.1).
Backward Elimination (Eliminación de Características): Después de agregar características, se eliminaron iterativamente las características menos significativas si el valor p asociado a esa característica era mayor o igual que el nivel de significancia de salida (SL_out, también establecido en 0.1).
Resultados: Las características seleccionadas utilizando Stepwise Selection para predecir el número de personas asesinadas en tiroteos en escuelas fueron 'casualties' (total de víctimas), 'injured' (número de personas heridas) y 'two_or_more' (indicador de si hubo dos o más atacantes).

En resumen, mediante el método de Stepwise Selection, se identificaron estas tres características como las más relevantes para predecir el número de personas asesinadas en tiroteos escolares. Este enfoque proporciona una selección eficiente y automatizada de características para análisis posteriores y puede ser útil para comprender mejor los factores que influyen en la gravedad de los tiroteos en entornos escolares.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


# Para escuelas privadas
X_privadas = tiroteos_privadas_por_año.index.values.reshape(-1, 1)
y_privadas = tiroteos_privadas_por_año.values

# Aplicar transformación logarítmica a las características
X_privadas_log = np.log(X_privadas)

X_train_privadas, X_test_privadas, y_train_privadas, y_test_privadas = train_test_split(X_privadas_log, y_privadas, test_size=0.5, random_state=42)

model_privadas = LinearRegression()
model_privadas.fit(X_train_privadas, y_train_privadas)
y_pred_privadas = model_privadas.predict(X_test_privadas)

mse_privadas = mean_squared_error(y_test_privadas, y_pred_privadas)
r2_privadas = r2_score(y_test_privadas, y_pred_privadas)

print("Escuelas Privadas:")
print("Error Cuadrático Medio (MSE):", mse_privadas)
print("Coeficiente de Determinación (R^2):", r2_privadas)

# Para escuelas públicas
X_publicas = tiroteos_publicas_por_año.index.values.reshape(-1, 1)
y_publicas = tiroteos_publicas_por_año.values

# Aplicar transformación logarítmica a las características
X_publicas_log = np.log(X_publicas)

X_train_publicas, X_test_publicas, y_train_publicas, y_test_publicas = train_test_split(X_publicas_log, y_publicas, test_size=0.5, random_state=42)

model_publicas = LinearRegression()
model_publicas.fit(X_train_publicas, y_train_publicas)
y_pred_publicas = model_publicas.predict(X_test_publicas)

mse_publicas = mean_squared_error(y_test_publicas, y_pred_publicas)
r2_publicas = r2_score(y_test_publicas, y_pred_publicas)

print("\nEscuelas Públicas:")
print("Error Cuadrático Medio (MSE):", mse_publicas)
print("Coeficiente de Determinación (R^2):", r2_publicas)

```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Supongamos que 'X' es tu característica 'Año'
X_privadas = tiroteos_privadas_por_año.index.values.reshape(-1, 1)
y_privadas = tiroteos_privadas_por_año.values

X_publicas = tiroteos_publicas_por_año.index.values.reshape(-1, 1)
y_publicas = tiroteos_publicas_por_año.values

# Aplicar transformación cuadrática
poly_features = PolynomialFeatures(degree=2)
X_privadas_poly = poly_features.fit_transform(X_privadas)
X_publicas_poly = poly_features.fit_transform(X_publicas)

# Modelo de regresión cuadrática
model_privadas_quad = LinearRegression()
model_publicas_quad = LinearRegression()

# Ajustar el modelo a los datos
model_privadas_quad.fit(X_privadas_poly, y_privadas)
model_publicas_quad.fit(X_publicas_poly, y_publicas)

# Predecir con el modelo cuadrático
y_pred_privadas_quad = model_privadas_quad.predict(X_privadas_poly)
y_pred_publicas_quad = model_publicas_quad.predict(X_publicas_poly)

# Graficar resultados para escuelas privadas
plt.figure(figsize=(10, 6))
plt.scatter(X_privadas, y_privadas, color='blue', label='Datos reales')
plt.plot(X_privadas, y_pred_privadas_quad, color='red', label='Regresión Cuadrática')
plt.title('Regresión Cuadrática para Escuelas Privadas')
plt.xlabel('Año')
plt.ylabel('Número de Tiroteos')
plt.legend()
plt.show()

# Graficar resultados para escuelas públicas
plt.figure(figsize=(10, 6))
plt.scatter(X_publicas, y_publicas, color='blue', label='Datos reales')
plt.plot(X_publicas, y_pred_publicas_quad, color='red', label='Regresión Cuadrática')
plt.title('Regresión Cuadrática para Escuelas Públicas')
plt.xlabel('Año')
plt.ylabel('Número de Tiroteos')
plt.legend()
plt.show()
```

```python
# Calcular el R^2
r2_privadas_quad = r2_score(y_privadas, y_pred_privadas_quad)
r2_publicas_quad = r2_score(y_publicas, y_pred_publicas_quad)

# Mostrar el R^2
print(f'R^2 para Escuelas Privadas: {r2_privadas_quad}')
print(f'R^2 para Escuelas Públicas: {r2_publicas_quad}')

```

Es un valor relativamente bajo, indicando que la relación cuadrática no explica completamente la variabilidad en los datos.

Para las escuelas públicas, el R^2 es aproximadamente 0.59. Esto significa que alrededor del 59% de la variabilidad en el número de tiroteos en escuelas públicas puede explicarse por la relación cuadrática con el año. Aunque es más alto que en el caso de escuelas privadas, aún hay una proporción significativa de variabilidad no explicada.

En resumen, los modelos de regresión cuadrática tienen un ajuste limitado a los datos.
